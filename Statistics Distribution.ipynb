```python
#1st binimial dirstribution
sample = np.random.binomial(10,0.5,size=10000)  # 1000 figures in binomial distribution 
plt.hist(sample, bins=12, align='left', density=True, rwidth=0.1)  # Hist 

plt.title('Binomial PMF with n={}, p={}'.format(10,0.5))  
plt.xlabel('number of successes')
plt.ylabel('probability')
```

```python
fig = plt.figure(figsize=(12,4.5)) #figure format
p1 = fig.add_subplot(121)
sample_1 = np.random.binomial(10,0.2,size=10000)  # 1000 figures in binomial distribution 
plt.hist(sample_1, bins=12, align='left', density=True, rwidth=0.1)  # subplot 1

p2 = fig.add_subplot(122)  # subplot 2
sample_2 = np.random.binomial(10,0.8,size=10000)  # 1000 figures in binomial distribution 
plt.hist(sample_2, bins=12, align='left', density=True, rwidth=0.1)
plt.show()
```

```python
#2nd poisson distribution 
lamb = 6
sample = np.random.poisson(lamb, size=10000)  # 10000 random poisson numbers
bins = np.arange(20)
plt.hist(sample, bins=bins, align='left', rwidth=0.1, density=True) # Hist

plt.title('Poisson PMF (lambda=6)')
plt.xlabel('number of arrivals')
plt.ylabel('probability')
plt.show()
```

```python
#3rd exponential distribution
tau = 10
sample = np.random.exponential(tau, size=10000)  # Genrating 1000 random figures
plt.hist(sample, bins=80, alpha=0.7, density=True) # hist
plt.margins(0.02) 

# Density distribution 
lam = 1 / tau
x = np.arange(0,80,0.1)
y = lam * np.exp(- lam * x)
plt.plot(x,y,color='orange', lw=3)

#Plot setting 
plt.title('Exponential distribution, 1/lambda=10')
plt.xlabel('time')
plt.ylabel('PDF')
plt.show()
```

```python
# 4th narmal distribution 
def norm_pdf(x,mu,sigma):
    '''Normal distribution'''
    pdf = np.exp(-((x - mu)**2) / (2* sigma**2)) / (sigma * np.sqrt(2*np.pi))
    return pdf

mu = 0    # average = o
sigma = 1 # std = 1

# generate figures
sample = np.random.normal(mu, sigma, size=10000)
plt. hist(sample, bins=100, alpha=0.7, density=True)

# ploting
x = np.arange(-5, 5, 0.01)
y = norm_pdf(x, mu, sigma)
plt.plot(x,y, color='orange', lw=3)
plt.show()
```

```python
height = df.height.dropna()
weight = df.weight.dropna()
```

```python
#height is normal distributed
fig = plt.figure(figsize=(16,5)) # picture srtting

p1 = fig.add_subplot(121)  # subplot
# PDF_1
std = height.std()
mean = height.mean()
x = np.arange(1.2, 2.2, 1/100)
y = norm_pdf(x, mean, std)
plt.plot(x,y, label='normal distribution')
# hist
plt.hist(height, bins=21, range=(1.2,2.2), rwidth=0.9, alpha=0.5, density=True, label='observables') #realdata 

plt.legend()
plt.xlabel('height')
plt.title('height_PDF')

p2 = fig.add_subplot(122)  #Subplot
# cdf
sample = np.random.normal(mean, std, size=10000)
plt.hist(sample, cumulative=True, bins=1000, range=(1.2,2.2), density=True, histtype='step', lw=2, label='normal distribution')
# ploting
plt.hist(height, cumulative=True, bins=1000, range=(1.2,2.2), density=True, histtype='step', lw=2, label='observables')# realdata 

plt.legend(loc='upper left')
plt.xlabel('height')
plt.title( 'height_CDF')
plt.show()
```

```python
#weight is distributed 
fig = plt.figure(figsize=(16,5)) # picture srtting

p3 = fig.add_subplot(121)  # subplot
# PDF_1
std = weight.std()
mean = weight.mean()
x = np.arange(0, 300, 300/100)
y = norm_pdf(x, mean, std)
plt.plot(x,y, label='normal distribution')
# hist
plt.hist(weight, bins=60, range=(0,300), rwidth=0.9, alpha=0.5, density=True, label='observables') #realdata 

plt.legend()
plt.xlabel('weight')
plt.title('weight_PDF')

p4 = fig.add_subplot(122)  #Subplot
# cdf
sample = np.random.normal(mean, std, size=10000)
plt.hist(sample, cumulative=True, bins=1000, range=(0,300), density=True, histtype='step', lw=2, label='normal distribution')
# ploting
plt.hist(weight, cumulative=True, bins=1000, range=(0,300), density=True, histtype='step', lw=2, label='observables')# realdata 

plt.legend(loc='upper left')
plt.xlabel('weight')
plt.title( 'weight_CDF')
plt.show()
```

```python
log_weight = np.log(weight)# log 
```

```python
#weight is distributed 
fig = plt.figure(figsize=(16,5)) # picture srtting

p3 = fig.add_subplot(121)  # subplot
# PDF_1
std = log_weight.std()
mean = log_weight.mean()
x = np.arange(3, 6, 3/100)
y = norm_pdf(x, mean, std)
plt.plot(x,y, label='normal distribution')
# hist
plt.hist(log_weight, bins=53, range=(3,6), rwidth=0.9, alpha=0.5, density=True, label='observables') #realdata 

plt.legend()
plt.xlabel('weight')
plt.title('weight_PDF')

p4 = fig.add_subplot(122)  #Subplot
# cdf
sample = np.random.normal(mean, std, size=10000)
plt.hist(sample, cumulative=True, bins=1000, range=(3,6), density=True, histtype='step', lw=2, label='normal distribution')
# ploting
plt.hist(log_weight, cumulative=True, bins=1000, range=(3,6), density=True, histtype='step', lw=2, label='observables')# realdata 

plt.legend(loc='upper left')
plt.xlabel('weight')
plt.title( 'weight_CDF')
plt.show()
```
